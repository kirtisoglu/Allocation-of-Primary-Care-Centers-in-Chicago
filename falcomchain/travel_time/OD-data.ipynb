```python
origin_df = pd.read_csv("/Users/kirtisoglu/Documents/GitHub/Chicago-Travel-Time-Matrix/network-data/origins.csv")
dest_df = pd.read_csv("/Users/kirtisoglu/Documents/GitHub/Chicago-Travel-Time-Matrix/network-data/destinations.csv")

origins = set(origin_df['id'])
destinations = set(dest_df['id'])

print(len(origins))
print(len(destinations))
```

```python
travel_times = {}
origin_to_dest = {}

for index, row in df.iterrows():
    
    origin = int(row['from_id'])
    destination = int(row['to_id'])
    time = int(row['travel_time_p50'])
    travel_times[(origin, destination)] = time
    
    if origin not in origin_to_dest.keys():
        origin_to_dest[origin] = set()
    
    origin_to_dest[origin].add(destination)
```

```python
missings = {}
for origin, value in origin_to_dest.items():
    missings[origin] = destinations - value
```

```python
for origin, dests in missings.items():
    
    #if len(missings[origin]) > 0:
        #print(missings[origin])
    
    if dests != {30185, 32588}:
        print(f"{origin} has the missing set {dests}")
```

```python
if 30185 in origins:
    print(30185)
if 32588 in origins:
    print(32588)
```

```python
for origin in origins:
    travel_times[(origin, 30185)] = 0
    travel_times[(origin, 32588)] = 0
```

```python
import pickle
path = "/Users/kirtisoglu/Documents/GitHub/Chicago-Travel-Time-Matrix/real_travel_times.pkl"
with open(path, "wb") as file:
    pickle.dump(travel_times, file)
```
